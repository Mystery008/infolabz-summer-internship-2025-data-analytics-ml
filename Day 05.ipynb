## 📈 What is Linear Regression?

Linear Regression is the **simplest form** of supervised learning used for predicting continuous numerical values.

### Types of Linear Regression
- **Simple Linear Regression**: 1 independent variable
- **Multiple Linear Regression**: 2 or more independent variables

We aim to find the **best-fit straight line** through the data points.



## 🧮 Mathematical Model

For **Simple Linear Regression**, the model predicts output \\( \hat{y} \\) using:

\[
\\( \hat{y} \\) = \\( \beta_0 \\)+ \\( \beta_1 \\) x
\]

Where:
- \\( \hat{y} \\): Predicted value (output)
- \\( x \\): Input feature
- \\( \beta_0 \\): Intercept (bias)
- \\( \beta_1 \\): Slope (weight)

### 📊 Graphical Interpretation

- The X-axis represents the input feature \\( x \\)
- The Y-axis represents the predicted output \\( \hat{y} \\)
- The best-fit line: \\( y = \beta_0 + \beta_1 x \\)
- Each point contributes a vertical error (called **residual**) to the line

````python
import matplotlib.pyplot as plt
import numpy as np

# Scatter plot
plt.scatter(df['area'], df['price'], label='Data Points')

# Fit a straight line (linear regression)
m, b = np.polyfit(df['area'], df['price'], 1)  # y = mx + b

# Plot the regression line
plt.plot(df['area'], m * df['area'] + b, color='red', label='Regression Line')

# Labels and legend
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Area vs Price')
plt.legend()
plt.show()
````

## 🧾 Summary: Understanding Linear Regression

This notebook presented the **mathematical theory** behind **Linear Regression**, a foundational algorithm in supervised machine learning. Below is a concise review of the key concepts covered:

### 🔍 What is Linear Regression?
- A method to model the relationship between a **dependent variable** (target) and an **independent variable** (feature).
- Used for **predicting continuous outcomes** based on input features.

### 🧠 Core Idea
Linear Regression assumes a **linear relationship** between the input variable \\( x \\) and the predicted output \\( \hat{y} \\):

\[
\\( \hat{y} \\) = \\( \beta_0 \\)+ \\( \beta_1 \\) x
\]

Where:
- \\( \hat{y} \\): Predicted value  
- \\( x \\): Input feature  
- \\( \beta_0 \\): Intercept (y-axis crossing point)  
- \\( \beta_1 \\): Slope (rate of change in y with respect to x)

### 📊 Visual Understanding

A straight line is drawn through the data points that best represents the overall trend.
